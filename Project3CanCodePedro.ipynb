In the following data base we will explore the correlations between different variables and the charges referring to Individual medical costs in dollars billed by health insurance in dollars. High charges of Individual medical costs billed by health insurance could be interpreted as indicator of deteriorated health. A short description of all columns included in this data set is illustrated in the following list:

•	age: Age of primary beneficiary in years.

•	sex: Insurance contractor gender, changed to 0 = female, 1 = male.

•	bmi: Body mass index, providing an understanding of body, weights that are relatively high or low relative to height, objective index of body weight (kg / m ^ 2) using the ratio of height to weight, ideally 18.5 to 24.9

•	children: Number of children covered by health insurance / Number of dependents.

•	smoker: Smoking (1 = yes, and 0 = no).

•	region: The beneficiary's residential area in the US, northeast, southeast, southwest, northwest.

•	charges: Individual medical costs billed by health insurance in dollars.




Hypothesis:

Is expected that older age, smoking habits, male gender, elevated body mass index, more childrens and metropolitan regions are associated to elevated medical costs. However some factor can have more influence than others, by calculating the correlation we can determine which factor have more impact/influence on medical charges.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load data from URL
df = pd.read_csv('https://raw.githubusercontent.com/PeEnAvRi/Project3CanCode/main/insurance.csv')

# Define the four BMI categories and their ranges
bins = [0, 18.5, 24.9, 29.9, 100]
labels = [' (0-18.5)', ' (18.5-24.9)', ' (24.9-29.9)', ' (30.0 or Above)']
df['BMI_category'] = pd.cut(df['bmi'], bins=bins, labels=labels)

# Calculate the mean and standard deviation of charges for each BMI category
charges_mean = df.groupby('BMI_category')['charges'].mean()
charges_std = df.groupby('BMI_category')['charges'].std()

# Define the colors for each BMI category
colors = ['purple', 'teal', 'orange', 'red']


# Create a bar plot of average charges by BMI category
fig, ax = plt.subplots()
bars = ax.bar(charges_mean.index, charges_mean, yerr=charges_std, align='center', alpha=0.5, ecolor='black', capsize=10)

# Add axis labels and a title
plt.xlabel('BMI Category')
plt.ylabel('Average Charges')
plt.title('BMI vs Average Charges')

# Add a legend with the BMI category ranges
legend_elements = [plt.Rectangle((0,0),1,1, color=colors[i]) for i in range(len(labels))]
plt.legend(legend_elements, labels, loc='best')

# Set the color of each bar to match the color in the legend
for i in range(len(bars)):
    bars[i].set_color(colors[i])

# Show the plot
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
#
# Load data from remote URL
df = pd.read_csv('https://raw.githubusercontent.com/PeEnAvRi/Project3CanCode/main/insurance.csv')

# Compute average BMI and standard deviation per age
bmi_by_age = df.groupby('age')['bmi'].agg(['mean', 'std']).reset_index()

# Plot average BMI with standard deviation vs age
plt.figure(figsize=(10, 6))
plt.scatter(bmi_by_age['age'], bmi_by_age['mean'], label='Average BMI')
plt.errorbar(bmi_by_age['age'], bmi_by_age['mean'], yerr=bmi_by_age['std'], fmt='none', capsize=3, label='Standard Deviation')
slope, intercept, _, _, _ = linregress(bmi_by_age['age'], bmi_by_age['mean'])
plt.plot(bmi_by_age['age'], intercept + slope * bmi_by_age['age'], color='red', label='Linear Regression')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.title('Average BMI with Standard Deviation vs Age')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from URL
df = pd.read_csv('https://raw.githubusercontent.com/PeEnAvRi/Project3CanCode/main/insurance.csv')

# Create histogram of charges
plt.hist(df['charges'], bins=5, color='blue')

# Add cumulative percentage line
total = float(len(df))
plt.ylim([0, total])
plt.twinx()
plt.plot(
    df['charges'].value_counts().sort_index().cumsum() / total * 100,
    color='red'
)

# Add labels and title
plt.title('Histogram with Pareto Chart of Charges')
plt.xlabel('Charges($)')
plt.ylabel('Frequency / Cumulative Percentage')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from URL
df = pd.read_csv('https://raw.githubusercontent.com/PeEnAvRi/Project3CanCode/main/insurance.csv')

# Define 5 charges categories
charge_categories = ['$0-$10,000', '$10,001-$20,000', '$20,001-$30,000', '$30,001-$40,000', '$40,001+']
charge_bins = [0, 10000, 20000, 30000, 40000, float('inf')]

# Add a new column 'charge_category' to the data frame
df['charge_category'] = pd.cut(df['charges'], bins=charge_bins, labels=charge_categories)

# Group the data by charge category and count the number of occurrences
charge_counts = df.groupby('charge_category').size().reset_index(name='count')

# Plot a pie chart of charge counts
plt.figure(figsize=(8, 8))
plt.pie(charge_counts['count'], labels=charge_counts['charge_category'], autopct='%1.1f%%')
plt.title('Charges by Category splited in 5 of $10,000')
plt.show()
```
