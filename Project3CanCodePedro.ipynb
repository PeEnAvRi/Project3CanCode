```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load data from remote URL
df = pd.read_csv('https://raw.githubusercontent.com/PeEnAvRi/Project3CanCode/main/insurance.csv')

# Create box plot of smoker vs charges
smoker_no = df[df['smoker'] == 'no']['charges']
smoker_yes = df[df['smoker'] == 'yes']['charges']

data = [smoker_no, smoker_yes]
labels = ['No', 'Yes']

fig, ax = plt.subplots()
ax.boxplot(data)
ax.set_xticklabels(labels)
plt.title('Smoker vs Charges')
plt.xlabel('Smoker')
plt.ylabel('Charges (in dollars)')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt 
from scipy.stats import linregress

# Load data from remote URL
df = pd.read_csv('https://raw.githubusercontent.com/PeEnAvRi/Project3CanCode/main/insurance.csv')

# Create scatter plot of age vs charges
plt.scatter(df['age'], df['charges'])
plt.title('Age vs Charges')
plt.xlabel('Age (years)')
plt.ylabel('Charges ($)')
plt.grid(True)

# Add linear regression line
slope, intercept, r_value, p_value, std_err = linregress(df['age'], df['charges'])
x = df['age']
y = slope * x + intercept
plt.plot(x, y, color='red')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from remote URL
df = pd.read_csv('https://raw.githubusercontent.com/PeEnAvRi/Project3CanCode/main/insurance.csv')

# Create histogram of charges
plt.hist(df['charges'], bins=8, color='blue')

# Add cumulative percentage line
total = float(len(df))
plt.ylim([0, total])
plt.twinx()
plt.plot(
    df['charges'].value_counts().sort_index().cumsum() / total * 100,
    color='red'
)

# Add labels and title
plt.title('Histogram with Pareto Chart of Charges')
plt.xlabel('Charges($)')
plt.ylabel('Frequency / Cumulative Percentage')
plt.show()
```
